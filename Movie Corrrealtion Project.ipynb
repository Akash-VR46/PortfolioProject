```python
# import libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure


%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# Read data

df = pd.read_csv(r'C:\Users\AKASH\Downloads\movies.csv')
```

```python
df.head()
```

```python
# Lets see if there is any empty data 

for col in df.columns:
    per_missing = np.mean(df[col].isnull())
    print('{}  - {}% '.format(col,per_missing))
```

```python
# datatypes for our column
df.dtypes
```

```python
# CHanging Datatypes

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
for i in range(len(df)):
    if pd.isnull(df['budget'][i]) or pd.isnull(df['gross'][i]):
        # print(i)
        df = df.drop(index=i)
```

```python
df.head()
# df['budget'][7664]
```

```python
df = pd.DataFrame.reindex(df)
```

```python
df
```

```python
# creating year corrected column
df['yearcorrected']= df['released'].astype(str).str[-20:-16]
```

```python
df = df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
# Drop the Duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
# df.drop_duplicates()
```

```python
# company high correlation
# budget high correlation
```

```python
# scatterplot budget vs gross
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earning')
plt.ylabel('Budget for film')
plt.show() 
```

```python
# plot the budget vs grosss

sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'red'},line_kws = {'color':'blue'})
```

```python
# Let's start looking at correlation
df.corr()  #pearson , kendall , spearman
```

```python
# High correlation between budget and gross

corr_matrix = df.corr(method='pearson')
sns.heatmap(corr_matrix,annot=True)
plt.title('Correaltion Matrix for Numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show() 
```

```python
df_numerized = df
df_numerized


for col_name in df_numerized.columns:
    if df_numerized[col_name].dtype == 'object':
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python

corr_matrix = df_numerized.corr(method='pearson')
sns.heatmap(corr_matrix,annot=True)
plt.title('Correaltion Matrix for Numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
corr_mat = df_numerized.corr()
corr_pairs = corr_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
